```python
import numpy as np

ans = np.random.randint(1, 101)  # Random number between 1 and 100

while True:
    a = int(input("Guess the number: "))

    if a == ans:
        print("Correct! You guessed the number.")
        break

    b = ans - a

    if b > 0:
        print("Too small!")
    else:
        print("Too big!")
```

```python
arr=np.random.randint(1,100,10) #start, stop, size
print(arr)
print(arr.shape)
print(arr.reshape(2,5)) #reshape the array to 2 rows and 5 columns
```

```python
#max min argmax argmin
print(arr)
print("Max:", arr.max())
print("Min:", arr.min())
print("Index of Max:", arr.argmax())
print("Index of Min:", arr.argmin())    
```

```python
#mean median std
print("Mean:", arr.mean())
print("Median:", np.median(arr))
print("Standard Deviation:", arr.std())
```
